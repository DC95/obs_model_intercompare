```python
    for epoch in range(start_epoch, config["training"]["epochs"]):
        start = time.time()
        model.train()
        running_loss = 0.0
        with logging_redirect_tqdm():
            for i, (conditions, targets) in enumerate(tqdm(data_loader, desc=f"Epoch {epoch+1}")):
                conditions, targets = conditions.to(device), targets.to(device)

                if conditions.ndim == 5:
                    B, C, T, H, W = conditions.shape
                    conditions = conditions.view(B, C * T, H, W)

                if targets.ndim == 5 and targets.shape[2] == 1:
                    targets = targets.view(targets.shape[0], targets.shape[1], targets.shape[3], targets.shape[4])
                t = torch.randint(0, timesteps, (targets.size(0),), device=device).long()
                loss = p_losses(model, targets, t, conditions, betas, standardize=True)

                optimizer.zero_grad()
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=config["training"]["clip_grad_norm"])
                optimizer.step()

                running_loss += loss.item()

                if writer:
                    global_step = epoch * len(data_loader) + i
                    writer.add_scalar("Loss/train", loss.item(), global_step)

            if scheduler:
                scheduler.step()
                if writer:
                    writer.add_scalar("LR", scheduler.get_last_lr()[0], epoch)

            avg_loss = running_loss / len(data_loader)
            logger.info(f"[Epoch {epoch}] Avg Loss: {avg_loss:.4f} | Time: {time.time() - start:.1f}s")

            if torch.cuda.is_available():
                mem = torch.cuda.memory_allocated() / 1e6
                logger.info(f"[Epoch {epoch}] GPU Mem: {mem:.1f} MB")

            if epoch % args.save_frequency == 0:
                checkpoint_path = os.path.join(args.output_path, f"ddpm_checkpoint_epoch_{epoch}.pt")
                torch.save({
                    "epoch": epoch,
                    "model_state_dict": model.state_dict(),
                    "optimizer_state_dict": optimizer.state_dict(),
                }, checkpoint_path)
                save_ddpm_sample(model, fixed_conditions, fixed_targets, os.path.join(args.output_path, "samples"), epoch, device, timesteps, betas)
                logger.info(f"[INFO] Saved model and sample for epoch {epoch}")
```
